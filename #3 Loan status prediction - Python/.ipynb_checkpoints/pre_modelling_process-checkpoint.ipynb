### Correlation analysis
Prior to the model fitting process, we must study the correlation between features. We must avoid including highly correlated features, first, because we do not want to include irrelevant information and, second, to avoid colinearity. Colinearity happens when a feature is dependent on another feature. This causes issues in model convergence due to singularities in the design matrix.

```python
# In order to calculate correlations between categorical features, we must recode the data

# Import methods to transform categorical features 
# Many methods do not work with object/string type features and thus these must be recoded
from sklearn.compose import ColumnTransformer
# Categorical features will be recoded using an ordinal encoder for further sequential feature selection
from sklearn.preprocessing import OrdinalEncoder


columns_to_keep_same = []
columns_to_keep_same.extend(numerical_features)
columns_to_keep_same.extend(['Loan_ID'])
columns_to_keep_same
```

```python
# Define a vector with the categorical features' names
categorical_features = df.columns.difference(columns_to_keep_same).tolist()
categorical_features
```

```python
# Create a ColumnTransformer to apply one-hot encoding to categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OrdinalEncoder(), categorical_features)
    ],
    remainder='passthrough'  # Pass through numerical features without any transformation
)

# The preprocessor will return the transformed data with a different column order from the original df
# Later I will turn the np.array into a pd.DataFrame and for that I need the right column order
import re
# First get the feature names ordered from the preprocessor
preprocessor_feature_names = preprocessor.fit(df).get_feature_names_out()

# Clean the "cat__" and "remainder__" prefixes
cleaned_feature_names = [re.sub(r'^(cat__|remainder__)', '', name) for name in preprocessor_feature_names]

print(cleaned_feature_names)
```

```python
# Transform the data
df = preprocessor.fit_transform(df)
df = pd.DataFrame(df, columns = cleaned_feature_names)

df
```

```python
# Last details
# I want Graduated = 1 and Not Graduated = 0
df["Education"].replace({0: 1, 1: 0}, inplace = True)
```

```python
# Do the same for normalised data
categorical_features = df_norm.columns.difference(columns_to_keep_same).tolist()

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OrdinalEncoder(), categorical_features)
    ],
    remainder='passthrough'  # Pass through numerical features without any transformation
)

preprocessor_feature_names = preprocessor.fit(df_norm).get_feature_names_out()
cleaned_feature_names = [re.sub(r'^(cat__|remainder__)', '', name) for name in preprocessor_feature_names]

df_norm = preprocessor.fit_transform(df_norm)

df_norm = pd.DataFrame(df_norm, columns = cleaned_feature_names)
df_norm["Education"].replace({0: 1, 1: 0}, inplace = True)
df_norm
```

### Feature Importance Analysis
Now that we have explored the correlations between numeric features, we are going to conduct a Feature Importance Analysis before fitting the models. 
This step will help us in determining which features are important or relevant to the loan status. It is a good practice that can save help us to save time in the model fitting process. 
To address this, we are going to use Random Forest methodologies and compute the Gini index, which determines the importance of the features towards the response.

#### Dividing the data
For the feature importance analysis and posterior model fitting process, we must split the data into train and test sets. The train set is to actually *train* the models and the test set is, of course, to *test* the model's performance. If we trained the model with the full dataset, the performance might be excellent, but when new data come the model will probably show a poor performance. Thus, an actual model's performance measure will be obtained with the test set. We will aim for 20% of the data for the testing set and 80% for the training set.


##### Some thoughts:
When it comes to providing loans, the actual risk for the business is to lend money to someone who will not pay (false positive). Losing a potential client who would actually pay, that is, rejecting the loan for a client who will pay (false negative) implies a cost of opportunity, but not a risk for the firm. Thus is preferable to lose a potential client rather than losing money. 

With that said, we must stress the model test and validation processes. The training process will comprise the validation process as well through k-fold cross-validation technique

The data set is comprised of 71% approved loands and 29% of rejected loans. This must be accounted for in the data splitting process. We cannot afford to have all approved loans in the training set. The result: all loans will be approved. We can assure that the train and test sets address these approved/rejected proportions using some specific arguments from a well known method

```python
# Import the train_test_split function from the model_selection module
from sklearn.model_selection import train_test_split

# First randomise the entire data 
df.sample(frac = 1, random_state = 25).reset_index(drop = True)

# Split approved loan data
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Loan_Status', 'Loan_ID'], axis = 1), df['Loan_Status'], 
                                                    test_size = 0.2, random_state = 25, stratify = df['Loan_Status']) 
# with stratify = df['Loan_Status'] assures that we keep 71% approved and 29% rejected for both train and test sets
```

```python
# Check if the approved/rejected proportions are consistent for train and test sets
(y_train.value_counts()/len(y_train), y_test.value_counts()/len(y_test))
# Perfect
```

```python
# Repeat the process for normalised data
df_norm.sample(frac = 1, random_state = 25).reset_index(drop = True)

X_n_train, X_n_test, y_n_train, y_n_test = train_test_split(df_norm.drop(['Loan_Status', 'Loan_ID'], axis = 1), 
                                                            df_norm['Loan_Status'], 
                                                            test_size = 0.2, random_state = 25, 
                                                            stratify = df_norm['Loan_Status']) 

print(len(X_n_train), len(X_n_test), len(y_n_train), len(y_n_test))
```

```python
# Lastly, recode the response 
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.fit_transform(y_test)

y_n_train = label_encoder.fit_transform(y_n_train)
y_n_test = label_encoder.fit_transform(y_n_test)
```

#### Random Forest Classifier to determine Feature Importance
As mentioned above it is key to determine the importance of the features in order to explain the response. In the current case the aim is to predict if a loan should be approved or rejected based on the applicant's features. We can address this using a Random Forest Classifier (RFC).
The main result from the RFC is *the average decrease of impurity across all trees within the model*. This process is exhibited in the **feature_importances_** attribute in the sklearn library. Then the outcomes are plotted and we can easily determine which features are important for the analysis (and which are not).

```python
# Conduct the feature importance analysis
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 500, random_state = 25)

forest.fit(X_train, y_train)
importances = forest.feature_importances_

indices = np.argsort(importances)[::-1]

plt.ylabel('Feature Importance')
plt.bar(range(X_train.shape[1]),
       importances[indices],
        align = 'center')

feat_labels = df.drop(['Loan_Status', 'Loan_ID'], axis = 1).columns
plt.xticks(range(X_train.shape[1]),
          feat_labels[indices], rotation = 90)
plt.xlim([-1, X_train.shape[1]])
plt.show()
```

Most important feature is Credit_History; given that the it supports the applicants background (or not). This feature is followed in importance by ApplicantIncome, LoanAmount,  and CoapplicantIncome. The applicant's income will determine their ability to pay; loan amount is certainly important and needs to make sense; and the coapplicant's income determines either an additional source of income to pay the loan or to back up the applicant's ability to pay.

Those features are then followed by the property area, loan amount term and the number of dependents, in fairly equal amounts of importance.

In the current case we have 381 unique entries and 11 features. Namely, we have plenty of data to *worry* about including all features due computational optimality and issues with model convergence. However, if we have irrelevant information, why should we include it? For instance, applicant's education, marital status, gender and self-employement status do not add relevant information to the analysis.
