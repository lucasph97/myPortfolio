# Loan status prediction
##### This dataset was obtained from [Kaggle](https://www.kaggle.com/datasets/bhavikjikadara/loan-status-prediction)

### About the loan_data.csv file:
* Loan_ID: A unique loan ID.
* Gender: Either male or female.
* Married: Weather Married(yes) or Not Marttied(No).
* Dependents: Number of persons depending on the client.
* Education: Applicant Education(Graduate or Undergraduate).
* Self_Employed: Self-employed (Yes/No).
* ApplicantIncome: Applicant income.
* CoapplicantIncome: Co-applicant income.
* LoanAmount: Loan amount in thousands.
* Loan_Amount_Term: Terms of the loan in months.
* Credit_History: Credit history meets guidelines.
* Property_Area: Applicants are living either Urban, Semi-Urban or Rural.
* Loan_Status: Loan approved (Y/N).

### Data imputation
There is little (if none) information on how to deal with missing values or what they might imply (e.g. does Credit_History = NaN means Credi_History = 0?). We will proceed on imputing the missing values by the mode, for categorical features, and median for numerical features

```python
loan_data.columns[loan_data.isnull().any()]
```

```python
for column in ['Gender', 'Dependents', 'Self_Employed','Credit_History']:
    loan_data[column].fillna(loan_data[column].mode()[0], inplace = True)
```

```python
loan_data['Loan_Amount_Term'].fillna(loan_data['Loan_Amount_Term'].median(), inplace = True)
```

## Descriptive analysis
In this section I will do a descriptive analysis. The aim is to obtain knowledge on the data composition using plots and metrics

```python
values = loan_data['Loan_Status'].value_counts() 
loan_statuses = values.index
loan_statuses = loan_statuses.map({'Y': 'Approved', 'N': 'Rejected'})

fig, ax = plt.subplots()
ax.pie(values, labels = loan_statuses, autopct='%1.1f%%')
ax.set(title = 'Percentage of Loans approved')
plt.show()
```

### Percentage of loans requested by gender and loan status
How many loans where requested by gender? How many of those were approved?

```python
values = loan_data['Gender'].value_counts() 
categories = values.index.tolist()

fig, ax = plt.subplots()
ax.pie(values, labels = categories, autopct='%1.1f%%')
ax.set(title = 'Percentage of Loans requested by Gender')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Loop across Gender categories
for i, gender in enumerate(loan_data['Gender'].unique()):
    # Filter the data by Gender
    filtered_data = loan_data[loan_data['Gender'] == gender]
    
    # Within the filtered data, obtain frequency for each gender 
    values = filtered_data['Loan_Status'].value_counts() 
    loan_statuses = values.index
    loan_statuses = loan_statuses.map({'Y': 'Approved', 'N': 'Rejected'})
        
    # Plot the pie chart in the corresponding subplot
    ax = axs[i]
    ax.pie(values, labels = loan_statuses, autopct='%1.1f%%')
    
    ax.set_title('Loan Status for ' + gender + ' applicants')
        
plt.tight_layout()
plt.show()
```

The percentage of approved loans is fairly similar for male and female applicants

### Percentage of loans requested by marital status and loan status

```python
values_married = loan_data['Married'].value_counts() 
categories_married = values_married.index
categories_married = categories_married.map({'Yes' : 'Married', 'No' : 'Not Married'})

fig, ax = plt.subplots()
ax.pie(values_married, labels = categories_married, autopct='%1.1f%%')
ax.set(title = 'Percentage of Loans requested by Marital Status')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Loop across Marital status categories
for i, married in enumerate(categories_married):
    # Filter the data by Gender
    if married == 'Married':
        filtered_data = loan_data[loan_data['Married'] == 'Yes']
    else:
        filtered_data = loan_data[loan_data['Married'] == 'No']
    
    # Within the filtered data, obtain frequency for each gender 
    values = filtered_data['Loan_Status'].value_counts()
    # And the categories
    categories = values.index
    categories = categories.map({'Y' : 'Approved', 'N' : 'Rejected'})  
    
    # Plot the pie chart in the corresponding subplot
    ax = axs[i]
    ax.pie(values, labels = categories, autopct='%1.1f%%')
    
    ax.set_title('Loan Status for ' + married + ' applicants')
        
plt.tight_layout()
plt.show()
```

There seems to be an considerable difference between the percentage of approved loans for married applicants compared to not married applicants.This might be because there are two sources of income that can afford the loan. Will explore that later on.

### Percentage of loans requested by education and loan status

```python
values = loan_data['Education'].value_counts() 
categories = values.index.tolist()

fig, ax = plt.subplots()
ax.pie(values, labels = categories, autopct='%1.1f%%')
ax.set(title = 'Percentage of Loans requested by Graduation Status')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Loop across Education categories
for i, educ in enumerate(loan_data['Education'].unique()):
    # Filter the data by Gender
    filtered_data = loan_data[loan_data['Education'] == educ]
    
    # Within the filtered data, obtain frequency for graduated and non-graduated applicants 
    values = filtered_data['Loan_Status'].value_counts() 
    loan_statuses = values.index
    loan_statuses = loan_statuses.map({'Y': 'Approved', 'N': 'Rejected'})
        
    # Plot the pie chart in the corresponding subplot
    ax = axs[i]
    ax.pie(values, labels = loan_statuses, autopct='%1.1f%%')
    
    ax.set_title('Loan Status for ' + educ + 'd applicants')
        
plt.tight_layout()
plt.show()
```

For Education, it might be a similar case to Marital Status: applicants who are graduated might have a higher-paid job and thus are more likely to pay the loan than not graduated applicants. That is why the percentage of approved loans seems higher for graduated applicants

### Percentage of loans requested by self-employement status and loan status

```python
values = loan_data['Self_Employed'].value_counts() 
categories = values.index.tolist()

fig, ax = plt.subplots()
ax.pie(values, labels = categories, autopct='%1.1f%%')
ax.set(title = 'Percentage of Loans requested by Self Employement Status')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Loop across Self-employement status categories
for i, selfemp in enumerate(loan_data['Self_Employed'].unique()):
    # Filter the data by selfemployed status
    filtered_data = loan_data[loan_data['Self_Employed'] == selfemp]
    
    # Within the filtered data, obtain frequency for self-employed and employed applicants 
    values = filtered_data['Loan_Status'].value_counts() 
    loan_statuses = values.index
    loan_statuses = loan_statuses.map({'Y': 'Approved', 'N': 'Rejected'})
        
    # Plot the pie chart in the corresponding subplot
    ax = axs[i]
    ax.pie(values, labels = loan_statuses, autopct='%1.1f%%')
    
    if selfemp == 'Yes':
        ax.set_title('Loan Status for Self Employed applicants')
    else:
        ax.set_title('Loan Status for Employed applicants')

        
plt.tight_layout()
plt.show()
```

It seems like for employed applicants the percentage of approved loans is a bit higher than for self employed applicants. This may be due to the fact that employees are more likely to have a constant cashflow while applicants who are self employed might have fluctuations on their cashflows. Thus employees are more likely to pay the loan

### Percentage of loans requested by property area and loan status

```python
values = loan_data['Property_Area'].value_counts() 
categories = values.index.tolist()

fig, ax = plt.subplots()
ax.pie(values, labels = categories, autopct='%1.1f%%')
ax.set(title = 'Percentage of Loans requested by Property Area')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))

# Loop across property area categories
for i, prop in enumerate(loan_data['Property_Area'].unique()):
    # Filter the data by property area
    filtered_data = loan_data[loan_data['Property_Area'] == prop]
    
    # Within the filtered data, obtain frequency by property area
    values = filtered_data['Loan_Status'].value_counts() 
    loan_statuses = values.index
    loan_statuses = loan_statuses.map({'Y': 'Approved', 'N': 'Rejected'})
        
    # Plot the pie chart in the corresponding subplot
    ax = axs[i]
    ax.pie(values, labels = loan_statuses, autopct='%1.1f%%')
    
    ax.set_title('Loan Status for ' + prop + ' area')

        
plt.tight_layout()
plt.show()
```

Apparently applicants from semiurban areas are more likely to get the loans approved, followed by applicants from urban areas and lastly from rural areas. In semiurban areas perhaps it is more likely to build small houses, while it is quite more expensive to build in urban areas; lastly rural areas might require higher loans for higher projects. Let's have a look

### Percentage of loans requested by property area, loan amount and status

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 10), sharex=True, gridspec_kw={'hspace': 0.5})


# Loop across property area categories
for i, prop in enumerate(loan_data['Property_Area'].unique()):
    # Filter the data by property area
    filtered_data = loan_data[loan_data['Property_Area'] == prop]
    
    # Within the filtered data, obtain frequency by property area
    loan_data_approved = filtered_data[filtered_data['Loan_Status'] == 'Y']
    loan_data_rejected = filtered_data[filtered_data['Loan_Status'] == 'N']

    # Calculate the histogram values for Approved and Rejected loans
    approved_hist, bins = np.histogram(loan_data_approved['LoanAmount'], bins=15)
    rejected_hist, _ = np.histogram(loan_data_rejected['LoanAmount'], bins=15)

    # Calculate the total number of data points
    total_data_points = len(loan_data)

    # Normalize the histogram values to get percentages
    approved_percentage = approved_hist / total_data_points * 100
    rejected_percentage = rejected_hist / total_data_points * 100

    # Set the width of the bars
    bar_width = 3

    # Generate the x-axis positions for the bars
    x1 = bins[:-1]
    x2 = x1 + bar_width
    
    # Plot the pie chart in the corresponding subplot
    ax = axs[i]
    
    # Plotting the dodged bars for Approved and Rejected loans
    ax.bar(x1, approved_percentage, width=bar_width, label='Approved')
    ax.bar(x2, rejected_percentage, width=bar_width, label='Rejected')

    # Adding labels and title
    ax.set_xlabel('Loan Amount')
    ax.set_ylabel('Percentage of Loans Requested')
    ax.set_title('Histogram of Loan Amount by Loan Status for ' + prop + ' property area')

    # Add a legend
    ax.legend()

# Displaying the histogram
plt.show()
```

Loan amount distribution seems kind of consistent for the three property areas. People from rural areas might have lower income and thus less likely to pay loans. For larger loan amounts the rate of approval is much greater for semiurban area compared to urban area.
